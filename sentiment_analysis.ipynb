## Preprocessing and Modelling

To build the corpus from the 'verified_reviews' we perform the following - <br>
1. Replace any non alphabet characters with a space
2. Covert to lower case and split into words
3. Iterate over the individual words and if it is not a stopword then add the stemmed form of the word to the corpus

```python
corpus = []
stemmer = PorterStemmer()
for i in range(0, data.shape[0]):
  review = re.sub('[^a-zA-Z]', ' ', data.iloc[i]['verified_reviews'])
  review = review.lower().split()
  review = [stemmer.stem(word) for word in review if not word in STOPWORDS]
  review = ' '.join(review)
  corpus.append(review)
```

Using Count Vectorizer to create bag of words

```python
cv = CountVectorizer(max_features = 2500)

# STORING INDEPENDENT AND DEPENDENT VARIABLES IN X AND y
X = cv.fit_transform(corpus).toarray()
y = data['feedback'].values
```

```python
# SAVING THE COUNT VECTORIZER
pickle.dump(cv, open('Models/countVectorizer.pkl', 'wb'))
```

Splitting data into train and test set with 30% data for testing.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 15)

print(f"X train: {X_train.shape}")
print(f"y train: {y_train.shape}")
print(f"X test: {X_test.shape}")
print(f"y test: {y_test.shape}")
```

```python
print(f"X train max value: {X_train.max()}")
print(f"X test max value: {X_test.max()}")
```

We will scale X_train and X_test so that all values are between 0 and 1.

```python
scaler = MinMaxScaler()

X_train_scl = scaler.fit_transform(X_train)
X_test_scl = scaler.transform(X_test)
```

```python
# SAVING THE SCALER MODEL
pickle.dump(scaler, open('Models/scaler.pkl', 'wb'))
```

## Predictions using the trained models

```python
# Load the CountVectorizer and scaler
cv_loaded = pickle.load(open('Models/countVectorizer.pkl', 'rb'))
scaler_loaded = pickle.load(open('Models/scaler.pkl', 'rb'))
```

```python
# Load the trained models
model_rf_loaded = pickle.load(open('Models/model_rf.pkl', 'rb'))
model_xgb_loaded = pickle.load(open('Models/model_xgb.pkl', 'rb'))
model_dt_loaded = pickle.load(open('Models/model_dt.pkl', 'rb'))
```

```python
# Function to preprocess input text
def preprocess_text(text):
    stemmer = PorterStemmer()
    stop_words = set(stopwords.words('english'))
    text = re.sub('[^a-zA-Z]', ' ', text)
    text = text.lower().split()
    text = [stemmer.stem(word) for word in text if word not in stop_words]
    text = ' '.join(text)
    return text
```

```python
# Function to map numerical predictions to 'POSITIVE' or 'NEGATIVE'
def map_to_sentiment(prediction):
    if prediction == 1:
        return 'Positive'
    else:
        return 'Negative'
```

```python
# Function to predict sentiment using ensemble of all three models
def predict_sentiment(text):
    # Preprocess the input text
    processed_text = preprocess_text(text)
    # Transform the preprocessed text using the CountVectorizer
    text_transformed = cv_loaded.transform([processed_text]).toarray()
    # Scale the transformed text using the MinMaxScaler
    text_scaled = scaler_loaded.transform(text_transformed)
    # Make predictions using all three loaded models
    rf_prediction = model_rf_loaded.predict(text_scaled)[0]
    xgb_prediction = model_xgb_loaded.predict(text_scaled)[0]
    dt_prediction = model_dt_loaded.predict(text_scaled)[0]
    # Combine predictions and return the majority vote
    votes = [rf_prediction, xgb_prediction, dt_prediction]
    majority_vote = max(set(votes), key=votes.count)
    # Map numerical prediction to 'POSITIVE' or 'NEGATIVE'
    sentiment = map_to_sentiment(majority_vote)
    return sentiment
```

```python
# Example usage - 1
text_input = "This product is amazing! I love it."
prediction = predict_sentiment(text_input)
print("SENTIMENT PREDICTION:", prediction)
```

```python
# Example usage - 2
text_input = "Sound is terrible if u want good music too get a bose"
prediction = predict_sentiment(text_input)
print("SENTIMENT PREDICTION:", prediction)
```
